```python
# 15. WAP to find common elements between two lists
a = list(map(int,input("Enter First List = ").split()))
b = list(map(int,input("Enter Secound List = ").split()))
common = []
for i in a:
    for j in b:
        if i==j and i not in common:
            common.append(i)
print("Common Elements = ",common)
```

```python
# 16. WAP to flatten a nested list (1 level deep)
a = eval(input("Enter Nested List = "))
flat = []
for i in a :
    if type(i) == list:
        for j in i:
            flat.append(j)
    else:
            flat.append(i)
print("Flatter List = ",flat)
```

```python
# 17. WAP to find frequency of each element in a list
a = list(map(int, input("Enter list: ").split()))

freq = {}

for i in a:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print("Frequency of elements:")
for key in freq:
    print(key, ":", freq[key])
```

```python
# 18. WAP to find the longest increasing sublist

a = list(map(int, input("Enter list: ").split()))

longest = []
temp = []

for i in range(len(a)):
    if i == 0 or a[i] > a[i-1]:
        temp.append(a[i])
    else:
        if len(temp) > len(longest):
            longest = temp
        temp = [a[i]]

if len(temp) > len(longest):
    longest = temp

print("Longest increasing sublist:", longest)
```

```python
# 20. WAP to sort a list without using built-in sort()
a = list(map(int, input("Enter list: ").split()))
for i in range(len(a)):
    for j in range(len(a)):
        if a[i]<a[j]:
            a[i] , a[j] = a[j],a[i]
print("sorted list" , a)
```

```python
# 28. WAP to find the equilibrium index in a list
a = list(map(int, input("Enter list: ").split()))
for i in range (len(a)):
    left = sum(a[:i])
    right = sum(a[i+1:])
    if left == right:
        print("Equlibrium Index" , i)
        break
else :
    print("No equlibrim index")
```

```python
# 29. WAP to find the maximum product of two elements in a list
a = list(map(int, input("Enter list: ").split()))
max_product = a[0]*a[1]
for i in range (len(a)):
    for j in range(i+1,len(a)):
        if a[i]*a[j] > max_product:
            max_product = a[i]*a[j]
print("Maximum Product ",max_product)
        
```
