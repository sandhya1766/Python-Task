```python
# 12.WAP to find all pairs in a list whose sum is equal to a given number*
a = list(map(int,input("Enter List").split()))
target = int(input("Enter target = "))
for i in range(len(a)):
    for j in range(i+1,len(a)):
        if a[i]+a[j]==target:
            print(a[i],a[j])
```

```python
# 13 . 3. WAP to rotate a list to the right by k positions
a = list(map(int,input("Enter List =").split()))
k = int(input("Enter value of K ="))
for i in range(k):
    last = a.pop()
    a.insert(0,last)
print(a)
```

```python
# 14. WAP to remove all occurrences of a given element from a list
a = list(map(int,input("Enter List = ").split()))
x = int(input("Enter Value to be Removed = "))
while x in a:
    a.remove(x)
print("Updated List" , a)
```

```python
# 15. WAP to find common elements between two lists
a = list(map(int,input("Enter First List = ").split()))
b = list(map(int,input("Enter Secound List = ").split()))
common = []
for i in a:
    for j in b:
        if i==j and i not in common:
            common.append(i)
print("Common Elements = ",common)
```

```python
# 16. WAP to flatten a nested list (1 level deep)
a = eval(input("Enter Nested List = "))
flat = []
for i in a :
    if type(i) == list:
        for j in i:
            flat.append(j)
    else:
            flat.append(i)
print("Flatter List = ",flat)
```

```python
# 17. WAP to find frequency of each element in a list
a = list(map(int, input("Enter list: ").split()))

freq = {}

for i in a:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print("Frequency of elements:")
for key in freq:
    print(key, ":", freq[key])
```
