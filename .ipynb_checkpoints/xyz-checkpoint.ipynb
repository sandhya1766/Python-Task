```python
# WAP to print user details (name, age, city)
name = input("Enter your name = ")
age = int(input("Enter your Age = "))
city = input("Enter your City = ")
print("Name = " , name)
print("Age = ", age)
print("City = ",city)
```

```python
# WAP to take input and display it.
data = input("Enter your Data = ")
print("Display Data = ",data)
```

```python
# WAP to swap two numbers.
a = int(input("Enter First Number"))
b = int(input("Enter Secound Number "))
a,b = b,a
print("After Swapping a is = ",a)
print("After Swapping b is = ",b)
```

```python
# WAP to convert string to integer.
str = "123"
print(int(str))
```

```python
# WAP to find ASCII value of a character.
chr = input("Enter Character =")
ascii = ord(chr) # ord() convert character to ascii and chr() convert ascii to character
print("Ascii value of ",chr,"is",ascii)
```

```python
# WAP to format output using f-string
name = input("Enter your name = ")
age = int(input("Enter your age = "))
print(f" My Name is {name}, And Age is {age}")
```

```python
#  WAP Taking multiple integers in one line
a, b = map(int, input("Enter two numbers: ").split())

print("First number:", a)   #input().split() → splits input by space , map(int, ...) → converts values to integers
print("Second number:", b)
```

```python
# WAP to check memory address of variable
x = 17
print(f" Memory Address of {x} is",id(x))
```
