```python
# 17. WAP to find frequency of each element in a list
a = list(map(int, input("Enter list: ").split()))

freq = {}

for i in a:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print("Frequency of elements:")
for key in freq:
    print(key, ":", freq[key])
```

```python
# 18. WAP to find the longest increasing sublist

a = list(map(int, input("Enter list: ").split()))

longest = []
temp = []

for i in range(len(a)):
    if i == 0 or a[i] > a[i-1]:
        temp.append(a[i])
    else:
        if len(temp) > len(longest):
            longest = temp
        temp = [a[i]]

if len(temp) > len(longest):
    longest = temp

print("Longest increasing sublist:", longest)
```

```python
# 28. WAP to find the equilibrium index in a list
a = list(map(int, input("Enter list: ").split()))
for i in range (len(a)):
    left = sum(a[:i])
    right = sum(a[i+1:])
    if left == right:
        print("Equlibrium Index" , i)
        break
else :
    print("No equlibrim index")
```

```python
# 38. WAP to merge two sorted lists without using sort()
a = list(map(int,input("Enter First List = ").split()))
b = list(map(int,input("Enter Secound List = ").split()))
result = []
for i in a:
    result.append(i)
for i in b:
    result.append(i)
for i in range(len(result)):
    for j in range(len(result)):
        if result[i]<result[j]:
            result[i],result[j] = result[j],result[i]
print("Resulted List",result)
```

```python
# 39. *WAP to find maximum subarray sum in a list (Kadane’s Algorithm)
a = list(map(int,input("Enter List").split()))
max_sum = a[0]
for i in range(len(a)):
    current_sum = 0
    for j in range(i,len(a)):
        current_sum+=a[j]
        if current_sum>max_sum:
            max_sum = current_sum
print("Maximum Subarray Sum ",max_sum)
```

```python
# 42. WAP to find the intersection of multiple lists.
a = list(map(int, input("Enter first list: ").split()))
b = list(map(int, input("Enter second list: ").split()))
c = list(map(int, input("Enter third list: ").split()))
result = []
for i in a:
    if i in b and i in c and i not in result:
        result.append(i)
print("Intersection is ",result)
```

```python
# 43. WAP to find the longest consecutive sequence in a list.
a = list(map(int, input("Enter list: ").split()))

a = list(set(a))   # remove duplicates
a.sort()

longest = []
temp = [a[0]]

for i in range(1, len(a)):
    if a[i] == a[i-1] + 1:
        temp.append(a[i])
    else:
        if len(temp) > len(longest):
            longest = temp
        temp = [a[i]]

# last check
if len(temp) > len(longest):
    longest = temp

print("Longest consecutive sequence:", longest)
```

```python
# 45. WAP to rearrange a list in alternating positive and negative numbers.
a = list(map(int, input("Enter List").split()))
pos = []
neg = []
for i in a:
    if i >= 0:
        pos.append(i)
    else:
        neg.append(i)
result = []
n = min(len(pos), len(neg))
for i in range(n):
    result.append(pos[i])
    result.append(neg[i])

print(result)
```

```python
# 50. WAP to shuffle elements of a list randomly without using shuffle()
import random
a = list(map(int,input("Enter List").split()))
for i in range(len(a)):
    j = random.randint(0,len(a)-1)
    a[i],a[j] = a[j],a[i]
print("Suffled list",a)
```
